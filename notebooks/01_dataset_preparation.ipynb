### Import Libraries

```python
import pandas as pd
```

### Load Dataset

```python
# season 14
s14_game_information = pd.read_csv("../data/raw/Draft Pick MPL Indonesia S14 - game information.csv")
s14_ban_heroes = pd.read_csv("../data/raw/Draft Pick MPL Indonesia S14 - ban heroes.csv")
s14_pick_heroes = pd.read_csv("../data/raw/Draft Pick MPL Indonesia S14 - pick heroes.csv")
# season 15
s15_game_information = pd.read_csv("../data/raw/Draft Pick MPL Indonesia S15 - game information.csv")
s15_ban_heroes = pd.read_csv("../data/raw/Draft Pick MPL Indonesia S15 - ban heroes.csv")
s15_pick_heroes = pd.read_csv("../data/raw/Draft Pick MPL Indonesia S15 - pick heroes.csv")
```

### Merge Dataset

```python
def merge_season_data(game_info_df: pd.DataFrame, ban_df: pd.DataFrame, pick_df: pd.DataFrame):
  merged = game_info_df.merge(ban_df, on="game_id")
  merged = merged.merge(pick_df, on="game_id")
  return merged
```

```python
# season 14
s14 = merge_season_data(s14_game_information, s14_ban_heroes, s14_pick_heroes)
s14["season"] = 14
s14.head()
```

```python
# season 15
s15 = merge_season_data(s15_game_information, s15_ban_heroes, s15_pick_heroes)
s15["season"] = 15
s15.head()
```

```python
# all season
all_season = pd.concat([s14, s15], axis=0).reset_index(drop=True)
all_season["game_id"] = range(1, len(all_season)+1)
all_season.to_csv("../data/processed/all_season.csv", index=False)
all_season.info()
```

### Draft Steps

```python
def create_draft_steps(season_df: pd.DataFrame):
  column_steps = {
    'team_1_ban_1': 'step_1', 'team_2_ban_1': 'step_2', 'team_1_ban_2': 'step_3', 'team_2_ban_2': 'step_4',
    'team_1_ban_3': 'step_5', 'team_2_ban_3': 'step_6', 'team_1_pick_1': 'step_7', 'team_2_pick_1': 'step_8',
    'team_2_pick_2': 'step_9', 'team_1_pick_2': 'step_10', 'team_1_pick_3': 'step_11', 'team_2_pick_3': 'step_12',
    'team_2_ban_4': 'step_13', 'team_1_ban_4': 'step_14', 'team_2_ban_5': 'step_15', 'team_1_ban_5': 'step_16',
    'team_2_pick_4': 'step_17', 'team_1_pick_4': 'step_18', 'team_1_pick_5': 'step_19', 'team_2_pick_5': 'step_20'
  }

  column_order = ['game_id', 'date', 'team_1', 'team_2', 'game_number', 'team_winner', 'patch_version', 'season',
                  'step_1', 'step_2', 'step_3', 'step_4', 'step_5', 'step_6', 'step_7', 'step_8', 'step_9', 'step_10',
                  'step_11', 'step_12', 'step_13', 'step_14','step_15', 'step_16', 'step_17', 'step_18', 'step_19', 'step_20']
  
  draft_steps = season_df.rename(columns=column_steps)[column_order]
  return draft_steps
```

```python
# all season
all_season_draft_steps = create_draft_steps(all_season)
all_season_draft_steps.to_csv("../data/processed/all_season_draft_steps.csv", index=False)
all_season_draft_steps.info()
```
