```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import confusion_matrix, classification_report, top_k_accuracy_score
from sklearn.naive_bayes import BernoulliNB, CategoricalNB, GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = np.loadtxt('../data/features/step7_features.csv', delimiter=',')
y = np.loadtxt('../data/features/step7_label.csv', delimiter=',')

print('X shape:', X.shape)
print('y shape:', y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=5)

print('X Train shape:', X_train.shape)
print('X Test shape:', X_test.shape)
print('y Train shape:', y_train.shape)
print('y Test shape:', y_test.shape)
```

```python
rf = RandomForestClassifier(100)
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
bnb = BernoulliNB()
bnb.fit(X_train, y_train)
bnb.score(X_test, y_test)
```

```python
ab = AdaBoostClassifier()
ab.fit(X_train, y_train)
ab.score(X=X_test, y=y_test)
```

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)
gnb.score(X_test, y_test)
```

```python
knn = KNeighborsClassifier(128)
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```
